## 5. Evaluation

### a. Accuracy

```python
# Protected groups

acc_F = accuracy_per_group(ind_F, y_test, pred)
acc_M = accuracy_per_group(ind_M, y_test, pred)

acc_imd_low = accuracy_per_group(ind_imd_low, y_test, pred)
acc_imd_high = accuracy_per_group(ind_imd_high, y_test, pred)

print("Acc F:", round(acc_F, 3))
print("Acc M:", round(acc_M, 3))
print("Acc IMD low:", round(acc_imd_low, 3))
print("Acc IMD high:", round(acc_imd_high, 3))
```

```python
# Global

round(accuracy(y_test, pred), 3)
```

### b. Recall

```python
rec_F = recall_per_group(ind_F, y_test, pred)
rec_M = recall_per_group(ind_M, y_test, pred)

rec_imd_low = recall_per_group(ind_imd_low, y_test, pred)
rec_imd_high = recall_per_group(ind_imd_high, y_test, pred)

print("Recall F:", round(rec_F, 3))
print("Recall M:", round(rec_M, 3))
print("Recall IMD low:", round(rec_imd_low, 3))
print("Recall IMD high:", round(rec_imd_high, 3))
```

```python
round(recall(y_test, pred), 3)
```

### c. Precision

```python
prec_F = precision_per_group(ind_F, y_test, pred)
prec_M = precision_per_group(ind_M, y_test, pred)

prec_imd_low = precision_per_group(ind_imd_low, y_test, pred)
prec_imd_high = precision_per_group(ind_imd_high, y_test, pred)

print("Prec F:", round(prec_F, 3))
print("Prec M:", round(prec_M, 3))
print("Prec IMD low:", round(prec_imd_low, 3))
print("Prec IMD high:", round(prec_imd_high, 3))
```

```python
round(precision(y_test, pred), 3)
```

### d. Demographic parity

```python
dp_F = demog_parity_per_group(ind_F, y_test, pred)
dp_M = demog_parity_per_group(ind_M, y_test, pred)

dp_imd_low = demog_parity_per_group(ind_imd_low, y_test, pred)
dp_imd_high = demog_parity_per_group(ind_imd_high, y_test, pred)

print("DP F:", round(dp_F, 3))
print("DP M:", round(dp_M, 3))
print("DP IMD low:", round(dp_imd_low, 3))
print("DP IMD high:", round(dp_imd_high, 3))
```

```python
round(demographic_parity(y_test, pred), 3)
```
